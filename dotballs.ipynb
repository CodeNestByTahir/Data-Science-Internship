```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df = pd.read_csv(r"C:\Users\khati\OneDrive\Desktop\Program_stud\python\Data Science\datasets\dots_per_team_per_over_per_ball.csv")
```

```python
ttd = df.groupby("batting_team")["dots"].sum()
plt.bar(ttd.index, ttd.values)
plt.xticks(rotation=90)
plt.xlabel("team")
plt.ylabel("total dot balls")
plt.title("total dot balls by each team")
plt.show()
```

```python
dotsofovers = df.groupby("over")["dots"].mean()
plt.plot(dotsofovers.index, dotsofovers.values)
plt.xlabel("over")
plt.ylabel("avg dot balls")
plt.title("avg dot balls per over")
plt.show()
```

```python
plt.scatter(df["over"], df["dots"])
plt.xlabel("over")
plt.ylabel("dot balls")
plt.title("over vs dot balls")
plt.show()
```

```python
plt.hist(df["dots"], bins=20)
plt.xlabel("dot balls")
plt.ylabel("frequency")
plt.title("distribution of dot balls")
plt.show()
```

```python
td = df.groupby("batting_team")["dots"].sum()
plt.pie(td, labels=td.index, autopct="%1.1f%%")
plt.title("dot ball by each teams")
plt.show()
```
