```python
plt.hist(df["Delivery_Time"], bins=30)
plt.xlabel("delivery time ")
plt.ylabel("no of orders")
plt.title("delivery time distribution")
plt.show()
```

```python
latetime = df.groupby("Traffic")["Delivery_Time"].mean()
plt.bar(latetime.index, latetime.values)
plt.xlabel("traffic level")
plt.ylabel("avg traaffic time")
plt.title("traffic vs delivery")
plt.show()
```

```python
latetime1 = df.groupby("Weather")["Delivery_Time"].mean()
plt.bar(latetime1.index, latetime1.values)
plt.xlabel("weather")
plt.ylabel("avg delivery time")
plt.title("weather vs delivery")
plt.show()
```

```python
latetime3= df.groupby("Vehicle")["Delivery_Time"].mean()
plt.bar(latetime3.index, latetime3.values)
plt.xlabel("vehicle type")
plt.ylabel("avg delivery time")
plt.title("vehicle vs delivery")
plt.show()
```

```python
latetime4= df.groupby("Area")["Delivery_Time"].mean()
plt.bar(latetime4.index,latetime4.values)
plt.xlabel("area type")
plt.ylabel("avg delivery time")
plt.title("area vs delivery")
plt.show()
```

```python
plt.scatter(df["Agent_Rating"], df["Delivery_Time"])
plt.xlabel("Agent Rating")
plt.ylabel("deliverytime")
plt.title("agent rating vs dwelivery time")
plt.show()
```

```python
plt.scatter(df["Agent_Age"], df["Delivery_Time"])
plt.xlabel("agent age")
plt.ylabel("delivery time")
plt.title("agent age vs delivery time")
plt.show()
```

```python
categories= df.groupby("Category")["Delivery_Time"].mean()
plt.pie(categories, labels=categories.index, autopct="%1.1f%%")
plt.title("avg delvery time by category")
plt.show()
```
