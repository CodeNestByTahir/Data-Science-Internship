```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\khati\OneDrive\Desktop\Program_stud\python\Data Science\datasets\zomato_.csv")
print(df)
```

```python
df['Cost_Category'] = "Low"
df.loc[df['Average_Cost_for_Two'] > 500, 'Cost_Category'] = "Medium"
df.loc[df['Average_Cost_for_Two'] > 800, 'Cost_Category'] = "High"
print(df)
```
