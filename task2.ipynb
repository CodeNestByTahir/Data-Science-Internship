```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

batsman = pd.read_csv(r"C:\Users\khati\OneDrive\Desktop\Program_stud\python\Data Science\datasets\sharma-kohli.csv")

plt.plot(batsman['season'],batsman['RG Sharma'],color="b",linewidth=2,marker="+",markersize=5,label="Rohit Sharma")
plt.plot(batsman['season'],batsman['V Kohli'],color="r",linewidth=2,marker="*",markersize=5,label="Virat Kohli")
plt.title("Rohit vs Virat")
plt.xlabel("Seasons")
plt.ylabel("Total runs")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv(r"C:\Users\khati\OneDrive\Desktop\Program_stud\python\Data Science\datasets\batsman.csv")

initial = df.head()
last = df.tail()
plt.scatter(initial['average'],initial['strike_rate'])
plt.scatter(last['average'],last['strike_rate'])
plt.show()

plt.plot(df["batsman"], df["total_runs"])
plt.title("Total Runs by IPL Batsmen")
plt.xlabel("Batsman")
plt.ylabel("Total Runs")
plt.show()

plt.figure()
plt.scatter(df["batsman"], df["strike_rate"])
plt.title("Strike Rate of Batsmen")
plt.show()

```
